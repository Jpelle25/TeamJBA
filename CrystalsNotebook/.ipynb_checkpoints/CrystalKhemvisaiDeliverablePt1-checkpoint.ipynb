#### Calculate COVID-19 data trends for last week of the data. Are the cases increasing, decreasing, or stable? Each student chooses a state to analyze.

According to the data, Georgia appears to have general increasing trend of covid cases between 01-03-23 and 01-16-23.


```python

df = pd.read_csv('covid_confirmed_usafacts.csv')

#filter for GA only
df = df[df['State'] == 'GA']

#First 5 columns saved
df1 = df.iloc[0:,:5] 

#extract dadta from last 2 weeks
df1.to_csv('./df1.csv', index=False)
df2 = df.iloc[:,-14:]

#get the sum of the counties for the past 2 weeks
d_sum = df2.sum()

# plot the progression of the cases over past 2 weeks of GA
d_sum.plot(figsize=(20, 10))
```

#### Each student member creates notebooks to read the Enrichment data and displays them on a notebook.

```python
# read enrichment data

enrich1 = pd.read_excel('employment1.xlsx')
enrich1 = enrich1[enrich1['Area Type'] != 'Nation']
enrich1 = enrich1[enrich1['Area Type'] != 'MSA']


enrich2 = pd.read_excel('employment2.xlsx')
enrich2 = enrich2[enrich2['Area Type'] != 'Nation']
enrich2 = enrich2[enrich2['Area Type'] != 'MSA']

enrich1.rename(columns = {'January Employment':'Month 1', 'February Employment':'Month 2', 'March Employment':'Month 3'}, inplace = True)
enrich2.rename(columns = {'April Employment':'Month 1', 'May Employment':'Month 2', 'June Employment':'Month 3'}, inplace = True)



frames = [enrich1, enrich2]
enrich = pd.concat(frames)
print(enrich)
```

#### Each student member performs initial merges with the COVID-19 data using the variables in the Enrichment data.


```python
# read enrichment data


enrichment = enrich.merge(enrich2)
print(enrichment)
```
