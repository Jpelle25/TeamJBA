```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set some Pandas options
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 25)

from datetime import datetime
```

```python
segments = pd.read_csv("covid_confirmed_usafacts.csv")
```

```python
wisconsin = segments[segments.StateFIPS==55].iloc[:,-7:]
wisconsin.head()
```

```python
wisconsin_sum = wisconsin.sum(axis=0)
wisconsin_sum.head()
```

```python
fig = plt.figure(figsize = (17,10))
ax = fig.gca()
wisconsin_sum.plot(style='o-',title="Total Covid Cases in Wisconsin Over One Week",ylabel="Total Number of Cases",ax=ax)
```

As can be seen in the line graph above, the data is clearly trending upwards, and as such, it can be said that the cases are steadily increasing in the last week of data.
